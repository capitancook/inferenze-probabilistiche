```python
import random

# Numero di lanci simulati
num_lanci = 1000

# Inizializziamo i contatori per testa e croce
testa_count = 0
croce_count = 0

# Simuliamo i lanci di moneta
for _ in range(num_lanci):
    lancio = random.choice(['testa', 'croce'])
    if lancio == 'testa':
        testa_count += 1
    else:
        croce_count += 1

# Calcoliamo le probabilità
probabilita_testa = testa_count / num_lanci
probabilita_croce = croce_count / num_lanci

# Stampiamo i risultati
print(f"Probabilità di testa: {probabilita_testa}")
print(f"Probabilità di croce: {probabilita_croce}")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generiamo dati casuali con una distribuzione normale
np.random.seed(0)
dati_altezze = np.random.normal(loc=170, scale=10, size=1000)  # media=170, deviazione standard=10, 1000 campioni

# Stima della media e della deviazione standard della distribuzione
media_stimata = np.mean(dati_altezze)
deviazione_standard_stimata = np.std(dati_altezze)

# Creiamo un istogramma dei dati e sovrapponiamo la distribuzione stimata
plt.hist(dati_altezze, bins=30, density=True, alpha=0.6, color='g', label='Dati campione')
x = np.linspace(140, 200, 100)
pdf_stimata = norm.pdf(x, loc=media_stimata, scale=deviazione_standard_stimata)
plt.plot(x, pdf_stimata, 'k-', linewidth=2, label='Distribuzione stimata')
plt.xlabel('Altezza (cm)')
plt.ylabel('Densità di probabilità')
plt.legend()
plt.show()

# Stampiamo i parametri stimati della distribuzione
print(f"Media stimata: {media_stimata}")
print(f"Deviazione standard stimata: {deviazione_standard_stimata}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generiamo dati casuali per X e Y
np.random.seed(0)
n_samples = 1000
X = np.random.normal(0, 1, n_samples)  # Variabile casuale X
Y = 2 * X + np.random.normal(0, 1, n_samples)  # Variabile casuale Y dipendente da X

# Creiamo un grafico di dispersione per visualizzare la distribuzione congiunta di X e Y
plt.scatter(X, Y, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico di Dispersione tra X e Y')
plt.grid(True)

# Calcoliamo la probabilità condizionata di Y dato X
# Ad esempio, calcoliamo P(Y > 1 | X > 0)
cond_X = X > 0
cond_Y = Y > 1
prob_condizionata = np.sum(cond_X & cond_Y) / np.sum(cond_X)

print(f'Probabilità di Y > 1 dato X > 0: {prob_condizionata:.2f}')

# Mostrare il grafico
plt.show()
```
